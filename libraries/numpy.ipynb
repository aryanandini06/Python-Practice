```python
import numpy as np

## create array using numpy
## create 1D array

arr1=np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
## reshape 2D

arr2=np.array([1,2,3,4,5])
arr2.reshape(1,5)  #...one row n 5 columns
```

```python
## 2D array

arr2=np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(arr2)
print(arr2.shape)
```
